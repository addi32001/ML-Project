```python
# Define the custom ranking for each ordinal variable
cut_categories = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_categories = ['D','E','F','G','H','I','J']
clarity_categories = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1','IF']
```

# pipeline

1. imputer
2. encoding
3. scaling

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline = Pipeline(
    steps =[
        ('imputer',SimpleImputer(strategy = "median")),
        ('scaler',StandardScaler()),
    ]
)

cat_pipeline = Pipeline(
    steps =[
        ('imputer',SimpleImputer(strategy = "most_frequent")),
        ('encoding',OrdinalEncoder(categories = [cut_categories,color_categories,clarity_categories])),
    ]
)

preprocessor = ColumnTransformer(
    [
        ('numerical_pipeline',num_pipeline,numerical_column),
        ('categorical_pipeline',cat_pipeline,categorical_column)
    ]
)


```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.30,random_state=30)
```

```python
preprocessor.fit_transform(X_train)
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
```

```python
X_train
```

```python
X_test = pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_test
```

```python
from sklearn.linear_model import  LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
import numpy as np
def evaluate_model(true, predicted):
    """
    Calculates and returns MAE, RMSE, and R2 score.
    """
    mae = mean_absolute_error(true, predicted)
    # np.sqrt() is used because mean_squared_error returns MSE by default
    rmse = np.sqrt(mean_squared_error(true, predicted)) 
    r2_square = r2_score(true, predicted)
    
    return mae, rmse, r2_square
```

```python
## Train mutiple models

models = {
    'LinearRegression': LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'ElasticNet':ElasticNet(),
}
trained_model_list = []
model_list =[]
r2_list = []

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)

    # Make Predictions
    y_pred = model.predict(X_test)
    mae, rmse, r2_square = evaluate_model(y_test,y_pred)
    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])
    print("Model Training Performance")
    print("RMSE",rmse)
    print("MAE",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)

    print("="*35)
    print("\n")
    
```
